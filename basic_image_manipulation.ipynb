```python
# Read the image 
img_eagle = cv2.imread('/content/Eagle_in_Flight.jpg', cv2.IMREAD_COLOR)

plt.figure(figsize = (8, 8))
plt.imshow(img_eagle[:, :, ::-1]);
```

```python
# i am now crop of the middle of the image 
cropped_region = img_eagle[20:420, 150:600]

# swap color channel BGR -- R:G:B
plt.imshow(cropped_region[:, :, ::-1]);
```

```python
# I am resize the image 
resized_cropped_region_2x = cv2.resize(cropped_region, None, fx = 2, fy = 2)

# display the image
plt.imshow(resized_cropped_region_2x[:, :, ::-1]);
```

```python
# Save resize image to file system.
cv2.imwrite('/content/resized_cropped_region_2x.png', resized_cropped_region_2x)

# display the cropped and resized image.
Image(filename = 'resized_cropped_region_2x.png')
```

```python
# this time i resize image and specifying size 
desired_width = 400
desired_height = 200
dim = (desired_width, desired_height)

# resize the image.
resized_cropped_region = cv2.resize(cropped_region, dsize = dim, interpolation = cv2.INTER_AREA)
plt.figure(figsize = (6, 6))
plt.imshow(resized_cropped_region[:, :, ::-1]);
```

```python
# resized the image and most of all maintaion aspect ratio
desired_width = 200
aspect_ratio = cropped_region.shape[0] / cropped_region.shape[1]
desired_height = int(desired_width * aspect_ratio)
dim = (desired_width, desired_height)

# Resize the image.
resize_cropped_region = cv2.resize(cropped_region, dsize = dim, interpolation = cv2.INTER_AREA)
plt.figure(figsize = (6, 6))
plt.imshow(resize_cropped_region[:, :, ::-1]);
```

```python
#  i am flipping image 
img_eagle_flipped_horz = cv2.flip(img_eagle, 1)
img_eagle_flipped_vert = cv2.flip(img_eagle, 1)
img_eagle_flipped_both = cv2.flip(img_eagle, -1)

# Display the images.
plt.figure(figsize = [18, 5])
plt.subplot(141); plt.imshow(img_eagle_flipped_horz[:, :, ::-1])
plt.title('Horizontal Flip')

plt.subplot(142); plt.imshow(img_eagle_flipped_both[:, :, ::-1])
plt.title('Vertical Flip')

plt.subplot(143); plt.imshow(img_eagle_flipped_both[:, :, ::-1])
plt.title('Both Flipped')

plt.subplot(144); plt.imshow(img_eagle[:, :, ::-1])
plt.title('Orginal')
```

```python
# Flip the image three ways.
img_eagle_flipped_horz = cv2.flip(img_eagle, 1)
img_eagle_flipped_vert = cv2.flip(img_eagle, 0)
img_eagle_flipped_both = cv2.flip(img_eagle, -1)

# Dispay the images.
plt.figure(figsize = [18, 5])
plt.subplot(141); plt.imshow(img_eagle_flipped_horz[:, :, ::-1]) 
plt.title('Horizontal Flip')
plt.subplot(142); plt.imshow(img_eagle_flipped_vert[:, :, ::-1])
plt.title('Vertical Flip')
plt.subplot(143); plt.imshow(img_eagle_flipped_both[:, :, ::-1]) 
plt.title('Both Flipped')
plt.subplot(144); plt.imshow(img_eagle[:, :, ::-1])            
plt.title('Original');
```

```python
# read the image 
img = cv2.imread('/content/New_Zealand_Boat.jpg', cv2.IMREAD_COLOR)
plt.figure(figsize = [10, 10])
plt.imshow(img[:, :, ::-1]);
```

```python
# Crop the image to extract the region around the sale boat.
crop_boat_image = img [100:300, 200:500]


# Resize the image up by a factor of 2x.
# 
resize_image = cv2.resize(crop_boat_image, None, fx = 2, fy = 2)

# Flip the cropped/resized image horizontally.
#
flip_resized_image = cv2.flip(resize_image, 1)

# Display the final result.
# 
# change the image color 
plt.imshow(flip_resized_image[:, :, ::-1]);
plt.figure(figsize = [5, 5]);
```
