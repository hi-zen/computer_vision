# Croping image 

```python
img_eagle = cv2.imread('/content/Eagle_in_Flight.jpg', cv2.IMREAD_COLOR)
plt.figure(figsize=(8, 8))
plt.imshow(img_eagle[:, :, ::-1]);
```

```python
cropped_region = img_eagle[20:420, 150:600]
plt.imshow(cropped_region[:, :, ::-1]);
```

```python
# resize the image 
resized_cropped_region_2x = cv2.resize(cropped_region, None,fx =2, fy = 2 )

# display the resized image
plt.imshow(resized_cropped_region_2x[:, :, ::-1]);
```

```python
cv2.imwrite('/content/resized_cropped_region_2x.png', resized_cropped_region_2x)
# display the cropped and resized image.
Image(filename =  '/content/resized_cropped_region_2x.png')
```

```python
# set the desired size of the image 
desired_width = 400
desired_height = 200

dim = (desired_width, desired_height)

# resize the image
resized_cropped_region = cv2.resize(cropped_region, dsize=dim, interpolation= cv2.INTER_AREA)
plt.figure(figsize= (6, 6))
plt.imshow(resized_cropped_region[:, :, ::-1]);
```

```python
desired_width = 200
aspect_ratio = cropped_region.shape[0]/ cropped_region.shape[1]
desired_height = int(desired_width * aspect_ratio)
dim = (desired_width, desired_height)

# resized the image 
resized_cropped_region = cv2.resize(cropped_region, dsize = dim, interpolation = cv2.INTER_AREA)
plt.figure(figsize=(6, 6))
plt.imshow(resized_cropped_region[:, :, ::-1]);
```

```python
# Flip the images three ways 
img_eagle_flipped_horzinatl = cv2.flip(img_eagle, -1)
img_eagle_flipped_vertical = cv2.flip(img_eagle, 0)
img_eagle_flipped_both = cv2.flip(img_eagle, -1)


# display the images
plt.figure(figsize=[18, 5])
plt.subplot(141); plt.imshow(img_eagle_flipped_horzinatl[:, :, ::-1])
plt.title('Horizonatl flip')

plt.subplot(142); plt.imshow(img_eagle_flipped_vertical[:, :, ::-1])
plt.title('Vertical flip')

plt.subplot(143); plt.imshow(img_eagle_flipped_both[:, :, ::-1])
plt.title("Both flipped")

plt.subplot(144); plt.imshow(img_eagle[:, :, ::-1])
plt.title('Orginal');
```

```python
img = cv2.imread('/content/New_Zealand_Boat.jpg', cv2.IMREAD_COLOR)
plt.figure(figsize=[6, 6])
plt.imshow(img[:, :, ::-1]);
```

```python
# Crop the image to extract the region around the sale boat.
crop_boat_image = img[150:300, 200:500]
# plt.imshow(crop_boat_image[:, :, ::-1]);

# Resize the image up by a factor of 2x.
resized_image_2 = cv2.resize(crop_boat_image, None, fx=2, fy=2)
# plt.imshow(resized_image_2[:, :, ::-1]);

# Flip the cropped/resized image horizontally.
flip_horizonatl_image = cv2.flip(resized_image_2, 1)


# Display the final result.
plt.figure(figsize = [50, 30])
plt.subplot(141);plt.imshow(flip_horizonatl_image[:, :, ::-1])
plt.title("Horizonatl flip");
```
