```python
# read the image
apollo_rocket = cv2.imread('/content/Apollo-8-Launch.jpg', cv2.IMREAD_COLOR)

# display the orginal image
plt.figure(figsize=[9, 6])
plt.imshow(apollo_rocket);
print('Shape of the image is - ', apollo_rocket.shape)
```

## Define the color anontaion 

```python
yellow = (0, 255, 255)
red = (0, 0, 255)
magenta = (255, 0, 255)
green = (0, 255, 0)
```

```python
# make a copy orginal image 
image_line = apollo_rocket.copy()

# draw yellow line (using - cv2.line_8)
image_line = cv2.line(image_line,(0,445),(450, 465), yellow, thickness=3, lineType=cv2.LINE_8)


# display the annoted image 
plt.figure(figsize=[10,10])
plt.imshow(image_line[:, :, ::-1]);
```

```python
# and this time using cv2.LINE_AA
image_line_aa = apollo_rocket.copy()

# draw a yellow line (using cv2.line_AA)
image_line_aa = cv2.line(image_line_aa, (0, 445), (450, 465), yellow, thickness=3, lineType=cv2.LINE_AA)

# display the annoted image
plt.figure(figsize=[10, 10])
plt.imshow(image_line_aa[:, :, ::-1]);
```

```python
# make a copy of the orginal image
image_circle = apollo_rocket.copy()

# draw a red circle
image_circle = cv2.circle(image_circle,(195, 55), 20, red, thickness=2, lineType=cv2.LINE_AA)

# display the annoted image
plt.figure(figsize=[8, 8])
plt.imshow(image_circle[:, :, ::-1]);
```

```python
image_rectangle = apollo_rocket.copy()

# draw rectangle 
image_rectangle = cv2.rectangle(image_rectangle, (300, 150), (480, 420), magenta, thickness=3, lineType=cv2.LINE_8)

# display the annotated image
plt.figure(figsize=[8, 8])
plt.imshow(image_rectangle[:, :, ::-1]);
```

```python
# now it's time to add text our images 
image_text = apollo_rocket.copy()
# add text to the image.
text = 'Apollo rocket lunce SpaceX'
font_face = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 0.8
font_color = yellow
font_thickness = 1
image_text = cv2.putText (image_text, text, (45, 30), font_face, font_scale, font_color, font_thickness, cv2.LINE_AA)

# display the annted image 
plt.figure(figsize = [10, 10])
plt.imshow(image_text[:, :, ::-1]);
```
